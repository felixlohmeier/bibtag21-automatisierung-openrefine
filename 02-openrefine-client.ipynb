# Schritt 2: openrefine-client ausführen

Wenn der Server gestartet ist (dauert 10-15 Sekunden), dann können Sie hier mit den vorbereiteten Befehlen OpenRefine über die Kommandozeile bedienen.

```bash
openrefine-client --create schriftstellerinnen.csv
```

```bash
openrefine-client --apply history.json schriftstellerinnen
```

```bash
openrefine-client --export schriftstellerinnen
```

# Schritt 3: Ergebnisse im Browser prüfen

Prüfen Sie abschließend in der Benutzeroberfläche von OpenRefine das Ergebnis.

Öffnen Sie dazu den Button "OpenRefine Session" im Launcher rechts. Im Menüpunkt "Open Project" sollte ein Projekt "Clipboard" erscheinen, dass die transformierten und angereicherten Daten enthält.
